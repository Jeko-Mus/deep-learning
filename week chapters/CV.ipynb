```python
#create bounds

#for blue
b_lower_bound = (170, 100, 20)
b_upper_bound = (180, 255, 255)

#for green
g_lower_bound = (38,40, 20)
g_upper_bound = (85, 255,255)

#for blue and green
bg_lower_bound = (0, 40, 20)
bg_upper_bound = (85, 255, 255)

#for blue,green and red
bgr_lower_bound = ()
bgr_upper_bound =()
```

```python
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

b_mask = cv2.inRange(hsv_image, b_lower_bound,b_upper_bound)
g_mask = cv2.inRange(hsv_image, g_lower_bound,g_upper_bound)
bg_mask = cv2.inRange(hsv_image, bg_lower_bound,bg_upper_bound)
#bgr_mask = cv2.inRange(hsv_image, bgr_lower_bound,bgr_upper_bound)
```

```python
masked = image.copy()
masked[b_mask==0] = (0, 0, 0)
plt.imshow(masked)
```

```python
masked = image.copy()
masked[g_mask==0] = (0, 0, 0)
plt.imshow(masked)
```

```python
masked = image.copy()
masked[bg_mask==0] = (0, 0, 0)
plt.imshow(masked)
```
