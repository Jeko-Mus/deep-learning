```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Annotations

```python
squirrel = cv2.imread('img/squirrel.jpg')
```

```python
rgb_sqr = cv2.cvtColor(squirrel, cv2.COLOR_BGR2RGB)
plt.figure(figsize = (10,8))
plt.imshow(rgb_sqr)
```

```python
sqr_copy = rgb_sqr.copy()
plt.figure(figsize = (10,8))
cv2.rectangle(sqr_copy, (80,160), (160,80), (0,0,0), 1)
cv2.rectangle(sqr_copy, (80,75), (160,20), (0,0,255), 1)
cv2.rectangle(sqr_copy, (170,175), (260,10), (255,0,0), 1)
cv2.putText(sqr_copy, 'squirrel', (70, 15), cv2.FONT_HERSHEY_PLAIN, 1.5, (0,0,255))
cv2.putText(sqr_copy, 'apple', (85, 125), cv2.FONT_HERSHEY_PLAIN, 1.5, (0,0,0))
cv2.putText(sqr_copy, 'leaves', (180, 150), cv2.FONT_HERSHEY_PLAIN, 1.5, (255,0,0))
plt.imshow(sqr_copy)
```

### ellipse

```python
#create blank image and copies
beach = cv2.imread('img/beach2.jpg')
rgb_beach = cv2.cvtColor(beach, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_beach)
```

```python
first = rgb_beach.copy()
second = rgb_beach.copy()
third = rgb_beach.copy()
fourth = rgb_beach.copy()
fifth = rgb_beach.copy()
```

```python
#image = cv2.ellipse(image, center_coordinates, axesLength, angle,startAngle, endAngle, color, thickness)
red = cv2.ellipse(first,(500,300),(300,200),0, 0, 360, (255,0,0),10)
green = cv2.ellipse(second,(500,300),(300,200),0, 90, 360, (0,255,0),10)
blue = cv2.ellipse(third,(500,300),(300,200),0, 180, 360, (0,0,255),10)
grey = cv2.ellipse(fourth,(500,300),(300,200),0, 270, 360, (0,0,0),10)
filled = cv2.ellipse(fifth,(500,300),(300,200),0, 270, 360, (0,0,0),cv2.FILLED)
```

```python
fig = plt.figure()
plt.figure(figsize = (25,20))
plt.subplot(1, 5, 1)
plt.imshow(red)

plt.subplot(1, 5, 2)
plt.imshow(green)

plt.subplot(1, 5, 3)
plt.imshow(blue)

plt.subplot(1, 5, 4)
plt.imshow(grey)

plt.subplot(1, 5, 5)
plt.imshow(filled)

plt.show()
```

### QR code reader

```python
qr_code_img = cv2.imread('img/qrcode.png')
```

```python
rgb_qrcode = cv2.cvtColor(qr_code_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_qrcode)
```

```python
detector = cv2.QRCodeDetector()
data, bbox, straight_qrcode = detector.detectAndDecode(qr_code_img)
```

```python
if bbox is not None:
    print('QR detected')
```

```python
print(bbox)
```

```python
qrcode_copy = rgb_qrcode.copy()
plt.figure(figsize = (10,8))
cv2.rectangle(qrcode_copy, (372,241), (488,126), (0,0,255), 6)
plt.imshow(qrcode_copy)
```

```python
qrcode_cropped = qrcode_copy[126:241, 372:488]
plt.imshow(qrcode_cropped)
```
